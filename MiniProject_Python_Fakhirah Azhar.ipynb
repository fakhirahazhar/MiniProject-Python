```python
# Import Numpy and Pandas

import numpy as np
import pandas as pd
```

```python
# Pembuatan ID
ID = []

for i in range(40):
    ID.append(i+1)
```

```python
D = pd.Series(ID)

ID
```

```python
# Tugas

tugas = np.random.normal(85, 7, 40)
for i in range (40):
    print(f'Nilai tugas mahasiswa {i+1} adalah {tugas[i]:.2f}')
```

```python
# UTS

UTS = np.random.normal(60, 15, 40)
for i in range (40):
    print(f'Nilai UTS mahasiswa {i+1} adalah {UTS[i]:.2f}')
```

```python
# UAS

UAS = np.random.normal(75, 10, 40)
for i in range (40):
    print(f'Nilai UAS mahasiswa {i+1} adalah {UAS[i]:.2f}')
```

```python
# Nilai akhir

nilai_akhir = tugas * 0.2 + UTS * 0.4 + UAS * 0.4
nilai_akhir = pd.Series(nilai_akhir)
for i in range(40):
    print(f'Nilai akhir mahasiswa {i+1} adalah {nilai_akhir[i]:.2f}')
```

```python
# melakukan binning
# membuat fungsi 'keterangan' dengan parameternya 'nilai_akhir'

def keterangan(nilai_akhir):
    if 85<=nilai_akhir<=100:
        return'A'
    elif 80<=nilai_akhir<85:
        return'A-'
    elif 75<=nilai_akhir<80:
        return'B+'
    elif 70<=nilai_akhir<75:
        return'B'
    elif 65<=nilai_akhir<70:
        return'B-'
    elif 60<=nilai_akhir<65:
        return'C+'
    elif 55<=nilai_akhir<60:
        return'C'
    elif 50<=nilai_akhir<55:
        return'C-'
    elif 20<=nilai_akhir<50:
        return'D'
    else:
        return'E'
```

```python
# mengaplikasikan nilai_akhir kedalam fungsi 'keterangan'
keterangan = nilai_akhir.apply(keterangan)

for i in range(40):
    print(f'Keterangan mahasiswa {i+1} adalah {keterangan[i]}')
```

```python
# membuat DataFrame
Rekap_Nilai_Mahasiswa = pd.DataFrame({'ID Mahasiswa':ID,
                    'Tugas':tugas,
                    'UTS':UTS,
                    'UAS':UAS,
                    'Keterangan':keterangan})
Rekap_Nilai_Mahasiswa
```
